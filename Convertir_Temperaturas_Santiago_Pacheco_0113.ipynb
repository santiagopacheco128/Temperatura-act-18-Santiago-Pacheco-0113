## 1. Fórmulas de Conversión

Para convertir entre estas dos escalas, utilizamos las siguientes relaciones matemáticas:

* **De Celsius (°C) a Fahrenheit (°F):** Multiplicamos los grados Celsius por 1.8 (o 9/5) y sumamos 32.
    $$°F = (°C \times 1.8) + 32$$

* **De Fahrenheit (°F) a Celsius (°C):** Restamos 32 a los grados Fahrenheit y el resultado lo dividimos entre 1.8.
    $$°C = \frac{°F - 32}{1.8}$$

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir 25°C a °F (Temperatura ambiente agradable)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar:** $45 + 32 = 77$
* **Resultado:** 25°C equivalen a **77°F**.

### Ejemplo B: Convertir 98.6°F a °C (Temperatura corporal promedio)
1.  **Identificar la fórmula:** $°C = (°F - 32) / 1.8$
2.  **Sustituir el valor:** $°C = (98.6 - 32) / 1.8$
3.  **Restar:** $98.6 - 32 = 66.6$
4.  **Dividir:** $66.6 / 1.8 = 37$
* **Resultado:** 98.6°F equivalen a **37°C**.

## 3. Código en Python

Este programa utiliza funciones para separar la lógica de cálculo del flujo del menú, haciendo que el código sea limpio y organizado.

```python
def convertirCF():
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

def convertirFC():
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

def Menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir Celsius a Fahrenheit")
        print("2. Convertir Fahrenheit a Celsius")
        print("3. Salir")
        
        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

# Llamada principal para iniciar el programa
if __name__ == "__main__":
    Menu()
```

### Explicación del código:
* **`while True`:** Crea un bucle infinito que permite al usuario realizar varias conversiones sin tener que reiniciar el programa.
* **`float()`:** Convertimos el `input` (que siempre es texto) a número decimal para poder hacer cálculos.
* **`f-strings` y `{:.2f}`:** Se usan para imprimir el resultado de forma elegante, limitando los decimales a 2 para mayor claridad.
* **`break`:** Es la instrucción que "rompe" el bucle `while` cuando el usuario elige la opción 3, terminando así la ejecución.

````python
def convertirCF():
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

def convertirFC():
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

def Menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir Celsius a Fahrenheit")
        print("2. Convertir Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Santiago Pacheco NL = 43")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

# Llamada principal para iniciar el programa
if __name__ == "__main__":
    Menu()
````
